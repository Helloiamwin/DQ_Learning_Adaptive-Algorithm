```python
Iteration = []
Ave_Loss = []
Ave_Reward = []
Sum_Loss = []
Sum_Reward = []
step = 1000
for i in [99999, 179999]:
    with open("trained_models/save_test_17_12_{}.pickle".format(i),'rb') as f:
        data = pickle.load(f)
        print(len(data))
        #print(i)
    for y in range (0,len(data),step):
        #print(y)
        Iteration.append(np.average((np.array(data))[y:(y+step),0]))
        Ave_Loss.append(np.average((np.array(data))[y:(y+step),1]))
        Sum_Loss.append(np.sum((np.array(data))[y:(y+step),1]))
        Ave_Reward.append(np.average((np.array(data))[y:(y+step),3]))
        Sum_Reward.append(np.sum((np.array(data))[y:(y+step),3]))
```

```python
plt.figure(figsize=(20,5))
plt.plot(Iteration, Ave_Loss, label = "Loss")
#plt.plot(Iteration, Sum_Reward, label = "Reward")
#plt.plot(Iteration, Ave_Reward, label = "Reward")
plt.title("Loss Averange Graph")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20,5))
plt.plot(Iteration, Ave_Reward, label = "Reward")
#plt.plot(Iteration, Sum_Reward, label = "Reward")
#plt.plot(Iteration, Ave_Reward, label = "Reward")
plt.title("Reward Averange Graph")
plt.legend()
plt.show()
```
